```python
#Reversing the linked list
class node:
    def __init__(self,data):
        self.data=data
        self.next=None
class linkedlist():
    def __init__(self):
        self.head=None
    def insertnode(self,data):
        newnode=node(data)
        newnode.next=self.head
        self.head=newnode
    def printing(self):
        temp=self.head
        while(temp):
            print(temp.data,end=" ->")
            temp=temp.next
        print("Null")
    def insertpos(self,data,pos):
        newnode=node(data)
        currentnode=self.head
        curpos=0
        while(currentnode):
            if curpos==pos:
                prev.next=newnode
                newnode.next=currentnode
            prev=currentnode
            currentnode=currentnode.next
            curpos+=1
    def dell(self):
        lastnode=self.head
        while(lastnode.next!=None):
            prev=lastnode
            lastnode=lastnode.next
        prev.next=None
    def dellpos(self,pos):
        lastnode=self.head
        poss=0
        while (lastnode):
            if poss==pos:
                nextnode=lastnode.next
                prev.next=nextnode
            prev=lastnode
            lastnode=lastnode.next
            poss+=1
    def reversing(self):
        prev=None
        current=self.head
        while(current):
            nextnode=current.next
            current.next=prev
            prev=current
            current=nextnode
            
        self.head=prev            
        
a=linkedlist()
a.insertnode(2)
a.insertnode(7)
a.insertnode(4)
a.insertnode(6)
a.insertnode(11)
a.insertnode(28)
a.insertnode(67)
a.printing()
a.insertpos(5,2)
a.printing()
a.dell()
a.printing()
a.dellpos(3)
a.printing()
a.reversing()
a.printing()
```
